**3.Generate a heatmap to visualize the distribution of order statuses based on age and gender.**

```python
df.corr(numeric_only=True)["Age"]
```

```python
data=df[["Gender","Age","Output"]]
data.head()
```

```python
data["Order_count"]=data["Output"].apply(lambda x:1 if x=="Yes" else 0)
```

```python
heatmap_df=data.pivot_table(index="Age",columns="Gender",values="Order_count",aggfunc="sum").fillna(0)
```

```python
sns.heatmap(heatmap_df,annot=True)
plt.title("Order Distribution by Age and Gender")
plt.show()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
```

```python
y
```
